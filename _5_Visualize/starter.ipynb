```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.sin(x) + 0.5

plt.figure(figsize=(8, 6))

# Plot the two lines
plt.plot(x, y1, label='Line 1', color='blue', linestyle='-')
plt.plot(x, y2, label='Line 2', color='red', linestyle='--')

# Fill only where y2 > y1
plt.fill_between(x, y1, y2, where=(y2 > y1), color='green', alpha=0.3, label='y2 > y1')

# (Optional) Fill where y1 > y2 with a different color
plt.fill_between(x, y1, y2, where=(y1 > y2), color='red', alpha=0.3, label='y1 > y2')

# Add labels and legend
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Conditional Shading Between Two Lines')
plt.legend()
plt.grid(True)
plt.show()
```
