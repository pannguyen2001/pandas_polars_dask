```python
# total_civilian_non_institutional_population_in_state_area, total_civilian_labor_force_in_state_area, total_employment_in_state_area, total_unemployment_in_state_area is number, but in csv, it has ",", so remove it and castype string
columns_remove_extra_comma = [
    "total_civilian_non_institutional_population_in_state_area", "total_civilian_labor_force_in_state_area",
    "total_employment_in_state_area",
    "total_unemployment_in_state_area"
    ]
df_us_unemployment[columns_remove_extra_comma] = df_us_unemployment[columns_remove_extra_comma].apply(lambda x: x.str.replace(',', ''), axis=1).astype(int)
```

```python
df_us_unemployment.rename({
    "percent_of_labor_force_unemployed_in_state_area": "unemployed_rate"
}, axis = 1, inplace = True)
```

```python
df_us_unemployment.columns
```

```python
# mean unemployed rate per state througn time
unemployed_mean_through_time = df_us_unemployment.groupby(["state_area"]).agg({
    "unemployed_rate": "mean"
}).round(2)
# unemployed_mean_through_time
```

```python
unemployed_mean_through_time["mean"] = float(unemployed_mean_through_time.mean().round(2))
# unemployed_mean_through_time
```

```python
# unemployed_mean_through_time["is_upper"] = unemployed_mean_through_time["unemployed_rate"] > unemployed_mean_through_time["mean"]

unemployed_mean_through_time[unemployed_mean_through_time["unemployed_rate"] > unemployed_mean_through_time["mean"]]
```

```python
unemployed_mean_through_time[unemployed_mean_through_time["unemployed_rate"] <= unemployed_mean_through_time["mean"]]
```

```python
unemployed_mean_through_time = unemployed_mean_through_time.reset_index()
```

```python
np.nanmedian(unemployed_mean_through_time["unemployed_rate"])
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(20,10))
plt.bar(x=unemployed_mean_through_time["state_area"], height= unemployed_mean_through_time["unemployed_rate"])

plt.axhline(
  y=np.nanmean(unemployed_mean_through_time["unemployed_rate"]), 
  color = 'g',
  linestyle = '--', 
  linewidth = 2,
  label = "Mean"
)
plt.axhline(
  y=np.nanmedian(unemployed_mean_through_time["unemployed_rate"]), 
  color = 'r',
  linestyle = '-.', 
  linewidth = 2,
  label = "Median"
)

plt.title("Unemployeed rate per state area")
plt.xlabel("State area")
plt.ylabel("Unemployed rate")
plt.xticks(rotation=45, ha='right')

plt.show()
```
